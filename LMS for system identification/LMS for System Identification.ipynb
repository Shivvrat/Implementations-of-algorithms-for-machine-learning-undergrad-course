```python
import numpy as np
import random 
import math
from pyswarms.utils.functions.single_obj import rosenbrock_func
from operator import itemgetter
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
"""
LMS function used to update weights
"""
def LMS (error, inp, weight, learning_rate = 0.5):
    return weight + (2 * learning_rate * inp * error)
```

```python
# for constants 0.2, 0.7, 0.2 with alpha = 0.01
weights = [np.random.randn(),np.random.randn(),np.random.randn()]
#print(weights)
first = 0
last = 10
num1 = 10000
error = 0
errors = []
diff = (last - first)/num1
for each in np.linspace(first, last, num=num1):
    weights_new = []
    actual = 0.2 * each + 0.7 * (each - diff) + 0.2 * (each - 2 * diff)
    output = weights[0] * each + weights[1] * (each - diff) + weights[2] * (each - 2 * diff)
    #print(actual, output, weights, each)
    weights_new.append(LMS(actual-output, each, weights[0], 0.01))
    weights_new.append(LMS(actual-output, each-1 * diff, weights[1], 0.01))
    weights_new.append(LMS(actual-output, each-2 * diff, weights[2], 0.01))
    weights = weights_new
    error = abs(output - actual)
    errors.append(error)
    if abs(output - actual) <= 0.0000001:
        break
print(weights)
print(error)
```

```python
plt.plot(errors)
plt.ylabel('values')
plt.xlabel('iterations')
plt.show()
```
